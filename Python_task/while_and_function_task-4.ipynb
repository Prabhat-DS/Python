```python
# q16 : Write a fun which will take input as a dict and give me out as a list of all the values 
# even in case of 2 level nesting it should work .
```

```python
d= {"k1":"ascfsfd", "k2":"afsf","k3":242,"k4":{"k5":"dwas","k6":323}}
def dict_to_list(d):
    l=[]
    for i in d.values():
        if type(i) != dict:
            l.append(i)
        
        if type(i)== dict:
            for j in i.values():
                l.append(j)
    return l
```
